```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_validate , KFold
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv("/Users/kato/Downloads/train (1).csv" , index_col="Id")
test_df = pd.read_csv("/Users/kato/Downloads/test (1).csv" , index_col="Id")
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
train_df2 = train_df.select_dtypes(include="object")
train_df2.head()
```

```python
train_category = train_df2
for col in train_category:
    lbl = LabelEncoder()
    train_df2[col] = lbl.fit_transform(list(train_df2[col].values))
```

```python
train_category
```

```python
train_col_list = train_df2.columns.tolist()
train_col_list
```

```python
train_df3 = train_df.drop(train_col_list, axis=1)
train_df3
```

```python
train_df = pd.merge(train_df3 , train_category , on="Id", how="left")
train_df.head()
```

```python
train_df.info()
```

```python
test_df2 = test_df.select_dtypes(include="object")
test_df2.head()
```

```python
test_category = test_df2
for col in test_category:
    lbl = LabelEncoder()
    test_df2[col] = lbl.fit_transform(list(test_df2[col].values))
```

```python
test_category
```

```python
test_col_list = test_df2.columns.tolist()
test_col_list
```

```python
test_df3 = test_df.drop(test_col_list, axis=1)
test_df3
```

```python
test_df = pd.merge(test_df3 , test_category , on="Id", how="left")
test_df.head()
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
y = train_df["SalePrice"]
X = train_df.drop(columns=["SalePrice"])
```

```python
X.head()
```

```python
y.head()
```

```python
model = xgb.XGBRegressor(objective="reg:squarederror" , max_depth=10)
kf = KFold(n_splits=5 , shuffle=True)
```

```python
score = cross_validate(model , X , y , cv=kf , scoring=["neg_mean_absolute_error"],return_estimator=True)
score
```

```python
mean_score = np.mean([score["estimator"][i].feature_importances_ for i in range(5)], axis=0)
mean_score
```

```python
model.fit(X , y)
```

```python
pred = model.predict(test_df)
```

```python
pred
```

```python
pred = pd.DataFrame(pred)
```
