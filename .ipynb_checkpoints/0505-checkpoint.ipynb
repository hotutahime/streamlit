```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_validate , KFold
from xgboost import plot_importance
from matplotlib import pyplot as plt
```

```python
train_df = pd.read_csv("/Users/kato/Desktop/berry/train.csv")
test_df = pd.read_csv("/Users/kato/Desktop/berry/test.csv")
test_dfid = pd.read_csv("/Users/kato/Desktop/berry/test.csv")
```

```python
ID = test_dfid["id"]
```

```python
test_df = test_df.drop(columns="id")
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
X = train_df.drop(columns=["id" , "yield"], axis=1)
y = train_df[["yield"]]
```

```python
X
```

```python
model = xgb.XGBRegressor(objective="reg:squarederror" , max_depth=5)
kf = KFold(n_splits=5 , shuffle=True)
```

```python
score = cross_validate(model , X , y , cv=kf , scoring=["neg_mean_absolute_error"], return_estimator=True)
```

```python
score
```

```python
model.fit(X , y)
```

```python
prediction = model.predict(test_df)
prediction
```

```python
plot_importance(model)
plt.show()
```

```python
submission = pd.DataFrame({
    "id":ID,
    "yield":prediction
})
```

```python
submission
```
