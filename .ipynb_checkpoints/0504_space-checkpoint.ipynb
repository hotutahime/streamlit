```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_validate , KFold
from xgboost import plot_importance
from matplotlib import pyplot as plt
```

```python
train_df = pd.read_csv("/Users/kato/Desktop/space_titanic/train.csv")
test_df = pd.read_csv("/Users/kato/Desktop/space_titanic/test.csv")
test_dfa = pd.read_csv("/Users/kato/Desktop/space_titanic/test.csv")
```

```python
ID = test_dfa["PassengerId"]
```

```python
train_df
```

```python
train_df["family"] = train_df["PassengerId"].str[:4]
train_df.head()
```

```python
train_feature = ["HomePlanet" , "CryoSleep" , "Cabin" , "Destination" , "VIP" , "Transported"]
for col in train_feature:
    lbl = LabelEncoder()
    train_df[col] = lbl.fit_transform(list(train_df[col].values))
```

```python
train_df.head()
```

```python
train_df = train_df.drop(["PassengerId" , "Name"], axis=1)
train_df
```

```python
test_df["family"] = test_df["PassengerId"].str[:4]
test_df.head()
```

```python
test_feature = ["HomePlanet" , "CryoSleep" , "Cabin" , "Destination" , "VIP"]
for col in test_feature:
    lbl = LabelEncoder()
    test_df[col] = lbl.fit_transform(list(test_df[col].values))
```

```python
test_df = test_df.drop(["PassengerId" , "Name"], axis=1)
test_df
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
train_df = train_df.fillna(train_df.mean())
test_df = test_df.fillna(test_df.mean())
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
train_df["family"] = train_df["family"].astype(int)
```

```python
test_df["family"] = test_df["family"].astype(int)
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
X = train_df.drop(columns="Transported")
y = train_df["Transported"]
```

```python
X.head()
```

```python
y.head()
```

```python
model = xgb.XGBClassifier(objective="binary:logistic" , max_depth=8)
kf = KFold(n_splits=5 , shuffle=True)
```

```python
score = cross_validate(model , X , y , cv=kf , scoring=["accuracy"], return_estimator=True)
score
```

```python
score["test_accuracy"].mean()
```

```python
model.fit(X ,y)
```

```python
prediction = model.predict(test_df)
prediction
```

```python
plot_importance(model)
plt.show()
```

```python
submission = pd.DataFrame({
    "PassengerId":ID,
    "Transported":prediction
})
```

```python
submission
```

```python
submission.to_csv("/Users/kato/Desktop/space_titanic/0504.csv" , index=False)
```
