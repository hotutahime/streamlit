```python
from bs4 import BeautifulSoup
import urllib.error , os
import urllib.request
import time , datetime
import pandas as pd
import numpy as np
```

```python
kana_list = ["a" , "i" , "u" , "e" , "o",
             "ka" , "ki" , "ku" , "ke" , "ko",
             "sa" , "si" , "su" , "se" , "so",
             "ta" , "ti" , "tu" , "te" , "to",
             "na"  , "ni" , "nu" , "ne" , "no",
             "ha" , "hi" , "fu" , "he" , "ho",
             "ma" , "mi" , "mu" , "me" , "mo",
             "ya" , "yu" , "yo",
             "ra" , "ri" , "ru" , "re" , "ro",
             "wa"
 ]
```

```python
sex_list = ["man/otoko" , "woman/onna"]

url_list = ["https://babyname.ojaru.jp/{0}-{1}.htm".format(sex , kana) for sex in sex_list for kana in kana_list]
```

```python
print(len(url_list))
for url in url_list:
    print(url)
```

```python
req = urllib.request.Request(url_list[0])
```

```python
html = urllib.request.urlopen(req)
```

```python
soup = BeautifulSoup(html , "html.parser")
```

```python
soup
```

```python
name_list = soup.find_all(["dt" , "dd"])

for k , v in enumerate(name_list):
    print(v.get_text().replace(" " , ""))
```

```python
hira = []
kanzi = []
han = []
sleep_time_sec = 0.5

for k , v in enumerate(name_list):
    
    if k % 2 == 0:
        hira.append(v.get_text().replace(" " , ""))
    else:
        kanzi.append(v.get_text().replace(" " , ""))
        
        if "otoko" in url_list[0]:
            han.append("男")
        else:
            han.append("女")

df = pd.DataFrame({"漢字":kanzi , "ひらがな":hira , "性別":han})
df
```

```python
for i in range(len(df)):
    if len(df.loc[i , "ひらがな"]) == 0:
        df.loc[i , "ひらがな"] = df.loc[i-1 , "ひらがな"]
```

```python
hira = []
kanzi = []
han = []
sleep_time_sec = 0.5

for url in url_list:
    time.sleep(sleep_time_sec)
    req = urllib.request.Request(url)
    html = urllib.request.urlopen(req)
    soup = BeautifulSoup(html , "html.parser")
    name_list = soup.find_all(["dt" , "dd"])

    for k , v in enumerate(name_list):
        if k % 2 == 0:
            hira.append(v.get_text().replace(" " , ""))
        else:
            kanzi.append(v.get_text().replace(" " , ""))
            if "otoko" in url:
                han.append("男")
            else:
                han.append("女")

df = pd.DataFrame({"漢字":kanzi , "ひらがな":hira , "性別":han})
for i in range(len(df)):
    if len(df.loc[i , "ひらがな"]) == 0:
        df.loc[i , "ひらがな"] = df.loc[i-1 , "ひらがな"]
df
```
