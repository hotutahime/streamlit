```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv("/Users/kato/Documents/data (1).csv" , index_col="年月日" , parse_dates=True)
```

```python
df
```

```python
df = df.drop(columns="曜日")
df
```

```python
df_month = df["2014":"2018"].resample("M").mean()
df_month
```

```python
df_month_pred = df.loc["2019"].resample("M").mean()
df_month_pred
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
y = df_month["平均気温(℃)"]
x = df_month[["日照時間(時間)" , "平均風速(m/s)"]]
sarima_model = sm.tsa.SARIMAX(y, x, order=(1,1,1) , seasonal_order=(1,1,1,12))
result = sarima_model.fit()

result.summary()
```

```python
df_month
```

```python
bestpred = result.predict("2018-12" , "2024-12", exog=df_month_pred[["日照時間(時間)" , "平均風速(m/s)"]])
```
