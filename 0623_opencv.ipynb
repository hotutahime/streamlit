```python
import cv2
import pandas as pd

filepath = "/Users/kato/Downloads/video_459968940195185204-y7OQzYEy.MP4"
cap = cv2.VideoCapture(filepath)

avg = None
frameNo = 0
list_df = pd.DataFrame(columns=["FrameNo", "x", "y", "w", "h", "cx", "cy", "area"])
contour_count = 0
list = [0,0,0,0,0]

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    frame = cv2.resize(frame, dsize=None, fx=0.6, fy=0.6)
    frameNo = frameNo+1
    fps = cap.get(cv2.CAP_PROP_FPS)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    if avg is None:
        avg = gray.copy().astype("float")
        continue
    
    bin_frame = cv2.threshold(gray, 180, 255, cv2.THRESH_BINARY)[1]
    cv2.accumulateWeighted(gray, avg, 0.80)
    frameDelta = cv2.absdiff(gray, cv2.convertScaleAbs(avg))
    
    thresh = cv2.threshold(frameDelta, 10, 255, cv2.THRESH_BINARY)[1]
    
    contours, hierarchy = cv2.findContours(frameDelta.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    thresh_con = frame.copy()
    
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 1000:
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(thresh_con, (x,y), (x+w,y+h), (0,255,0), cv2.LINE_4)
            M = cv2.moments(contour)
            cx = int(M["m10"]/M["m00"])
            cy = int(M["m01"]/M["m00"])
            
            list.pop(0)
            list.append(frameNo)
            print(list)
            if list[1] - list[0] > 20:
                contour_count + 1
                
            a = pd.Series([frameNo, x, y, w, h, cx, cy, area], index=list_df.columns)
            list_df = list_df.append(a, ignore_index=True)
            
            cv2.putText(thresh_con,
                        text="area="+str(area),
                        org=(x-10, y-10),
                        fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                        fontScale=1,
                        color=(0,255,0),
                        thickness=2,
                        lineType=cv2.LINE_4)
            
    cv2.putText(thresh_con,
                text="count="+str(contour_count),
                org=(10,30),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=1,
                color=(0,255,0),
                thickness=2,
                lineType=cv2.LINE_4)
            
    cv2.imshow("Frame", frame)
    cv2.imshow("Frame2", thresh)
    cv2.imshow("Frame3", thresh_con)
    cv2.imshow("Frame4", frameDelta)
    key = cv2.waitKey(1)
    if key == 27:
        break
            
    cap.release()
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.waitKey(0)
        
# cap.release()
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# cv2.waitKey(0)
```
