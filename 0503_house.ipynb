```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_validate , KFold
from xgboost import plot_importance
from matplotlib import pyplot as plt
```

```python
train_df = pd.read_csv("/Users/kato/Desktop/house/train.csv" , index_col=0)
test_df = pd.read_csv("/Users/kato/Desktop/house/test.csv" , index_col=0)
test_dfa = pd.read_csv("/Users/kato/Desktop/house/test.csv")
```

```python
ID = test_dfa["Id"]
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
train_df2 = train_df.select_dtypes(include="object")
train_df2.head()
```

```python
train_category = train_df2
for i in train_category:
    lbl = LabelEncoder()
    train_df2[i] = lbl.fit_transform(list(train_df2[i].values))
```

```python
train_category
```

```python
train_col_list = train_df2.columns.tolist()
train_col_list
```

```python
train_df3 = train_df.drop(train_col_list, axis=1)
train_df3
```

```python
train_df4 = pd.merge(train_df3 , train_category, on="Id", how="left")
train_df4
```

```python
train_df4.head()
```

```python
train_df4.info()
```

```python
train_df5 = train_df4.fillna(train_df4.mean())
```

```python
train_df5.info()
```

```python
test_df2 = test_df.select_dtypes(include="object")
test_df2
```

```python
test_category = test_df2
for j in test_category:
    lbl = LabelEncoder()
    test_df2[j] = lbl.fit_transform(list(test_df2[j].values))
```

```python
test_category
```

```python
test_col_list = test_df2.columns.tolist()
```

```python
test_col_list
```

```python
test_df3 = test_df.drop(test_col_list, axis=1)
test_df3
```

```python
test_df4 = pd.merge(test_df3 , test_category, on="Id" , how="left")
test_df4
```

```python
test_df4.info()
```

```python
test_df5 = test_df4.fillna(test_df4.mean())
```

```python
test_df5.info()
```

```python
X = train_df4.drop(columns=["SalePrice"])
y = train_df4["SalePrice"]
```

```python
X
```

```python
y
```

```python
model = xgb.XGBRegressor(objective="reg:squarederror" , max_depth=10)
kf = KFold(n_splits=5 , shuffle=True)
```

```python
score = cross_validate(model , X , y , cv=kf , scoring=["neg_mean_absolute_error"], return_estimator=True)
```

```python
score
```

```python
model.fit(X , y)
```

```python
prediction = model.predict(test_df5)
prediction
```

```python
plot_importance(model)
plt.show()
```

```python
submission = pd.DataFrame({
    "Id":ID,
    "SalePrice":prediction
})
```

```python
submission
```

```python
submission.to_csv("/Users/kato/Desktop/house/0503_2house.csv" , index=False)
```
