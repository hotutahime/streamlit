```python
import pandas as pd
import numpy as np
```

```python
reserve_tb = pd.read_csv("/Users/kato/Desktop/mypython/awesomebook-master/data/reserve.csv" , encoding="utf-8")
reserve_tb.head(15)
```

```python
#指定列の抜き出しはdf名[["列名", "列名", "列名"......]])と記述する。
reserve_tb[["reserve_id","hotel_id","customer_id"]].head(3)
```

```python
#値を絞り込むにはquery関数を使う。関数なので、（）で囲む。ただし、中で文字列をダブルクオーテーションで囲う必要があるため、
#関数全体はシングルクオーテーションで囲む必要がある。
reserve_tb.query('"2016-10-13" <= checkout_date <= "2016-10-14"').head(3)
```

```python
#データ量が多すぎる場合はサンプル抽出する。
reserve_tb.sample(frac=0.5).head(3)
```

```python
target = pd.Series(reserve_tb["customer_id"].unique()).sample(frac=0.5)
target
```

```python
#isinはdf名[df名["列名"].isin
#dfの中で、検索列名の中で、一致したものを抜き出す
reserve_tb[reserve_tb["customer_id"].isin(target)].head(3)
```

```python
#resultとは、hotel_id毎に値を計算をしたく、計算するためのaggを呼び出し、reserve_idの個数をcount関数で数え、
#customer_idの重複を除いた数を数える
result = reserve_tb.groupby("hotel_id").agg({"reserve_id":"count" , "customer_id":"nunique"})
result
#もう少し詳しくすると、groupbyは指定の列の値毎に集約する。ピボットテーブルのID重複削除のようなイメージ。
#aggでは、列に計算方法を指定できる。
#例えばreserve_id:countなら、reserve_idの個数を数える。
#nuniqueは、重複を削除するという意味。
```

```python
#上のresultは指定したのでhotel_idがインデックスになっているが、
#変数に対してreset_index関数をかけるとインデックスを振り直せる。
#inplace=Trueとは、振り直したindex値を確定値として扱う（一時的なフリ直しではないということ）
result.reset_index(inplace=True)
result.columns = ['hotel_id', 'rsv_cnt', 'cus_cnt']
result
```

```python
result = reserve_tb.groupby('hotel_id').agg({'total_price': ['max', 'min', 'mean', 'median',lambda x: np.percentile(x, q=20)]}).reset_index()
result.columns = ['hotel_id', 'price_max', 'price_min', 'price_mean','price_median', 'price_20per']
#resultとは、reserve_tbの中のhotel_id毎に、total_priceの最高、最小、平均、中央値を出し、
```
