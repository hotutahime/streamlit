```python
import requests
from bs4 import BeautifulSoup
import urllib
import pandas as pd
```

```python
url = "https://travel.rakuten.co.jp/yado/aomori/aomori.html?lid=jparea_undated_map"
```

```python
html = requests.get(url)
html.encoding = html.apparent_encoding
soup = BeautifulSoup(html.content , "html.parser")
```

```python
hotel_price = []

for htlprice in soup.find_all(class_="htlPrice"):
    htlprice_text = htlprice.text.replace("\n","").replace("\t","")
    hotel_price.append(htlprice_text)
```

```python
hotel_price
```

```python
hotel_names = []

for element in soup.find_all("h1"):
    text = element.text.replace("\n","").replace("\t","").replace("こちらの宿泊施設は、宿泊者からの総合評価点数が高い、もしくは多くの宿泊実績がある等の独自の条件を満たしたプリファードプログラム参加施設です。楽天トラベルへサービス利用料を支払うことにより、原則同条件の他の施設よりも上位に表示されています。（ヘルプページ参照）","")
    hotel_names.append(text)
```

```python
hotel_names
```

```python
df = pd.DataFrame({"ホテル名": hotel_names})
df
```

```python
df2 = df.drop(df.index[0])
df2
```

```python
df3 = df2.reset_index(drop=True)
df3 = df3.drop(columns="index")
df3
```

```python
df4 = pd.DataFrame({"価格": hotel_price})
df4
```

```python
df5 = pd.concat([df3, df4], axis=1)
df5
```
