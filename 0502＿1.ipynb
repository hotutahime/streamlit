```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_validate , KFold
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("/Users/kato/Downloads/train.csv")
```

```python
train.head()
```

```python
features = ["Sex","Embarked"]

for col in features:
    lbl = LabelEncoder()
    train[col] = lbl.fit_transform(list(train[col].values))
```

```python
train.head()
```

```python
traindr = train.drop(["Name","Ticket","Cabin"], axis=1)
```

```python
traindr.head()
```

```python
traindr.info()
```

```python
X = traindr.drop(columns=["Survived"])
y = traindr["Survived"]
```

```python
X.head()
```

```python
y.head()
```

```python
model = xgb.XGBClassifier(objective="binary:logistic" , max_depth=100)
kf = KFold(n_splits=5, shuffle=True)
score = cross_validate(model , X , y , cv=kf, scoring=["accuracy"], return_estimator=True)
```

```python
score
```

```python
test = pd.read_csv("/Users/kato/Downloads/test.csv")
```

```python
test = test.drop(["Name","Ticket","Cabin"], axis=1)
```

```python
test2 = ["Sex","Embarked"]

for col in test2:
    lbl = LabelEncoder()
    test[col] = lbl.fit_transform(list(test[col].values))
```

```python
model.fit(X, y)
```

```python
pred = model.predict(test)
```
