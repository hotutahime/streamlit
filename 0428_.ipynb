```python
import pandas as pd
import numpy as np
```

```python
datasets = pd.read_csv("/Users/kato/Downloads/Boston (1).csv")
test_data = pd.read_csv("/Users/kato/Downloads/Boston (1)-2.csv")
```

```python
datasets.head()
```

```python
test_data.head()
```

```python
y = datasets["medv"]
X = datasets[["crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","black","lstat"]]

print(y.shape)
print(X.shape)
```

```python
test = test_data[["crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","black","lstat"]]
```

```python
test.head()
```

```python
import xgboost as xgb
from sklearn.model_selection import cross_validate, KFold
model_XGB = xgb.XGBRegressor(objective="reg:squarederror", max_depth=10)
kf = KFold(n_splits=10 , shuffle=True)
```

```python
score = cross_validate(model_XGB, X , y , cv=kf, scoring=["neg_mean_absolute_error"], return_estimator=True)
```

```python
score
```

```python
score["test_neg_mean_absolute_error"].mean()
```

```python
mean_score = np.mean([score["estimator"][i].feature_importances_ for i in range(5)],axis=0)
```

```python
mean_score
```

```python
model_XGB.fit(X,y)
```

```python
pred = model_XGB.predict(test)
```
