```python
import pandas as pd
import numpy as np
```

```python
#csvを読み込む
reserve_tb = pd.read_csv("/Users/kato/Desktop/python/taizen/data/reserve.csv" , encoding="utf-8")
```

```python
#読み込んだCSVの最初の五行だけを表示する
reserve_tb.head()
```

```python
#読み込んだCSVの特定の列だけを表示するにはdf[["列名（複数可）"]]
reserve_tb[["reserve_id",
           "hotel_id",
           "customer_id",
           "reserve_datetime",
           "checkin_date",
           "checkin_time",
           "checkout_date"]]
```

```python
#query関数で特定の値のデータを抽出する
reserve_tb.query('"2016-10-13" <= checkout_date <= "2016-10-14"')
```

```python
#抽出したデータ量が多すぎる場合はランダムサンプリングでデータ量をランダムに減らす
#sampleメソッド＋タプル＋frac=0〜1（0％〜100％）
#ただし、これでは全体ランダムサンプリングなのでどこかの列を基準に等ではない
reserve_tb.sample(frac=0.5)
```

```python
#target変数にはreserve_tbのcustomer_idを一旦Series（sampleはSeriesじゃないと使えない）にし、
#そこからuniqueメソッドで重複を削除した上でcustomer_idの全体の50％をtarget変数に格納する
#isinで、target変数にある50％のIDと一致したIDのみを抜き出す＝実質IDでサンプリングした事と一緒
target = pd.Series(reserve_tb["customer_id"].unique()).sample(frac=0.5)
reserve_tb[reserve_tb["customer_id"].isin(target)]
```

```python
#agg関数で、辞書型によって集計処理をまとめて設定できる
#reserve_tbの中の、hotel_id毎に、reserve_idの個数カウントと、customer_idの重複を除いた個数をカウント
#count=個数、nunique=重複を除いた個数
#（一行で示すなら）reserve_tbを対象に、hotel_idの、reserve_idの個数と、customer_idの重複を除いた個数を集計した
result = reserve_tb.groupby("hotel_id").agg({"reserve_id":"count", "customer_id":"nunique"})
result
```

```python
#df.reset_indexで行番号を再設定し、inplace=Trueで上書きをした
#df.columns=["列名"]で、それぞれの列名を変更した
result.reset_index(inplace=True)
result.columns = ["hotel_id" , "rsv_cnt" , "cus_cnt"]
result
```

```python
#reserve_tbの、hotel_idとpeople_numの２つ毎に（リストで指定）、
 #これはhotel_idのh1でもpeople_numが複数あるため、それを分けるイメージ
#total_priceの値の合算値（sum関数）を計算し、最後に行番号を振り直した
result = reserve_tb.groupby(["hotel_id", "people_num"])["total_price"].sum().reset_index()
result
```

```python
#この方法でも列名を変更可（辞書の中は古い列名：変更する新しい列名となっている）
#inplace=Trueは上書きするということ
result.rename(columns={"total_price":"price_sum"}, inplace=True)
result
```

```python
#reserve_tbの、hotel_id毎に、aggで複数の条件を指定するがここではリストで複数指定している
#指定しているのは最大値、最低値、平均値、中央値、パーセンタイル値
#パーセンタイル値とは、値の大きさで並び替えた時に、指定したパーセントの順位にいる値のこと
#20%なら、小さい方から数えて20％の順位に位置する値

#lambda式とは、「lambda x:」が呪文、その後にxを処理する関数を書く記述式のこと
#この例では、np.percentile(x, q=20)なので、xに対しnumpyのパーセンタイル関数を使用してq=20なので
#xの20（ここではaggでtotal_priceを指定しているので、total_priceの20％という意味になる）
#の値を出してねという意味
result = reserve_tb.groupby("hotel_id").agg({"total_price":["max","min","mean","median", lambda x: np.percentile(x, q=20)]}).reset_index()
result
```

```python
#列名の変更
result.columns = ["hotel_id", "price_max", "price_min", "price_mean", "price_median", "price_20per"]
result
```

```python
#reserve_tbの、hotel_idを集約基準に設定して、
#aggで集約処理を複数してする、ここではリストを使用して
#total_priceをvar=分散値、stb=標準偏差値で集計し、
#最後に行番号を振り直している
result = reserve_tb.groupby("hotel_id").agg({"total_price":["var","std"]}).reset_index()
result
```

```python
#列名の変更
#fillnaは欠損値があった場合はそれを0にする
result.columns = ["hotel_id", "price_var", "price_std"]
result.fillna(0, inplace=True)
result
```

```python
#一番よく出てくる最頻値の計算
#mode関数で一発で出来る
#ただし、round関数で小数点（n=引数の位）を指定している
reserve_tb["total_price"].round(-3).mode()
```

```python
#total_priceの中で一番よく出てくる値
reserve_tb["total_price"].mode()
```

```python
#rank関数で並び替える前準備として、reserve_datetimeのデータ型をタイムスタンプ型に変更
#理由はrank関数が文字型に対応していないため（見た目は変わらないが、object→timestanpに変更している）
reserve_tb["reserve_datetime"] = pd.to_datetime(reserve_tb["reserve_datetime"], format="%Y-%m-%d %H:%M:%S")
reserve_tb
```

```python
#log_noとは、reserve_tbのcustomer_id毎に、
#reserve_datetimeが、ascending=Trueによって照準（早い予約日時順）に並び替えをしている
#medhod="first"は、同じ値が現れた場合に、登場順にする（重複させない）
reserve_tb["log_no"] = reserve_tb.groupby("customer_id")["reserve_datetime"].rank(ascending=True, method="first")
reserve_tb
```

```python
#hotel_id（ホテル毎）毎に、size関数で予約個数を取得（size関数については要調査）
rsv_cnt_tb = reserve_tb.groupby("hotel_id").size().reset_index()
rsv_cnt_tb
```

```python
#列名の変更
rsv_cnt_tb.columns = ["hotel_id", "rsv_cnt"]
rsv_cnt_tb
```

```python
#rsv_cnt_rank列とは、rankを降順
rsv_cnt_tb["rsv_cnt_rank"] = rsv_cnt_tb["rsv_cnt"].rank(ascending=False, method="min")
rsv_cnt_tb
```

```python
#不要列を削除するだけ
rsv_cnt_tb.drop("rsv_cnt", axis=1, inplace=True)
rsv_cnt_tb
```

```python
hotel_tb = pd.read_csv("/Users/kato/Desktop/python/taizen/data/hotel.csv", encoding="utf-8")

#reserve_tbのquery関数でpeople_numが1の行だけを抽出し、
#それとhotel_tbもquery関数でis_businessの列だけを抽出し、
#両方のDFにあるhotel_idを基準に（くっつける時の参照値のようなもの）、
#inner結合は要調査
pd.merge(reserve_tb.query("people_num == 1"), hotel_tb.query("is_business"), on="hotel_id", how="inner")
```

```python
#hotel_tbには、大きいエリアの名前と、小さいエリアの名前の二つが設定されている
hotel_tb
```

```python
import gc

#hotel_tbのbig_area_nameとsmall_area_nameを集計軸とし、
#それぞれの要素数（ここでは個数？）を取得している（size関数）
#as_index=Falseは、groupbyを使うとデフォルトでグループラベル名がindexになるため、
#それを避けるために行番号を取得するためのもの
#as_index=Falseを指定しないと、big_area_nameがindexになるイメージ（実際はキーエラーになった）
#infoを使うとわかるが、キーエラーになるのは多分、タイプがobjectのため集計できないということ。
small_area_mst = hotel_tb.groupby(["big_area_name", "small_area_name"], as_index=False).size().reset_index()
small_area_mst = small_area_mst[["big_area_name", "small_area_name", "size"]]
small_area_mst
#small_area_mst.info()
```

```python
#p101まで（23・6・21）
```

```python
small_area_mst.columns = ["big_area_name", "small_area_name", "hotel_cnt"]
small_area_mst
```

```python
#自分のホテルを見た方に対し、別のホテルを紹介する前処理
#自分のホテルを引くので、全ホテルー1をしていて、
#個数が20以上ならbigを表示し、20以下ならsmallを表示している
small_area_mst["join_area_id"] = np.where(small_area_mst["hotel_cnt"] -1 >= 20, small_area_mst["small_area_name"],small_area_mst["big_area_name"])
small_area_mst
```

```python
#不要なhotel_cnt, big_area_name列を削除
small_area_mst.drop(["hotel_cnt", "big_area_name"], axis=1, inplace=True)
small_area_mst
```

```python
base_hotel_mst = pd.merge(hotel_tb, small_area_mst, on="small_area_name").loc[:, ["hotel_id", "join_area_id"]]
base_hotel_mst
```

```python
#concatで、small_area_nameのhotel_idと、
#big_area_nameのhotel_idを縦に結合している（と思う）
#更に、結合するときに列名をrenameメソッドで辞書を使い変更しながら結合している
recommend_hotel_mst = \
pd.concat([hotel_tb[["small_area_name", "hotel_id"]]\
           .rename(columns={"small_area_name": "join_area_id"},
                   inplace=False),hotel_tb[["big_area_name", "hotel_id"]]\
           .rename(columns={"big_area_name": "join_area_id"}, inplace=False)])
recommend_hotel_mst
```

```python
recommend_hotel_mst.rename(columns={"hotel_id": "rec_hotel_id"}, inplace=True)
recommend_hotel_mst
```

```python
pd.merge(base_hotel_mst, recommend_hotel_mst, on="join_area_id")\
.loc[:, ["hotel_id", "rec_hotel_id"]]\
.query("hotel_id != rec_hotel_id")
```

```python
#過去データ結合を行うために（過去の顧客毎の利用価格帯を調べるために過去データを探る）
#customer_id毎にgroupbyで集約を行うが、
#applyによってgroup毎にreserve_datetimeの並び替えを行う
result = reserve_tb.groupby("customer_id").apply(lambda group: group.sort_values(by="reserve_datetime", axis=0, inplace=False))
result
```

```python
#過去の価格の列を追加
#total_priceだけをcustomer_id毎に集計し、
#shift perios=2で下に二行ずらしている理由は
#顧客IDの過去2回前の予約金額を持ってきているため
result["before_price"] = pd.Series(result["total_price"].groupby("customer_id").shift(periods=2))
result
```

```python
#reserve_datetimeを並び替え
result = reserve_tb.groupby("customer_id")\
.apply(lambda x: x.sort_values(by="reserve_datetime", ascending=True))\
.reset_index(drop=True)
result
```

```python
#新たな列price_avgとは
#customer_id毎にtotal_priceのwindowを3件にまとめ、その平均値を計算
#min_periods=1は、1件以上あった場合に計算する設定
result["price_avg"] = pd.Series(result.groupby("customer_id")\
                               ["total_price"].rolling(center=False, window=3, min_periods=1).mean().reset_index(drop=True))
result
```

```python
result["price_avg"] = result.groupby("customer_id")["price_avg"].shift(periods=1)
result
```
