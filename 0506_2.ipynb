```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://travel.rakuten.co.jp/yado/aomori/aomori.html?lid=jparea_undated_map"
html = requests.get(url)
html.encoding = html.apparent_encoding
```

```python
soup = BeautifulSoup(html.content , "html.parser")
soup
```

```python
hotel = soup.find_all("a")
hotel
```

```python
price = soup.find_all("p" , attrs={"class":"htlPrice"})
price
```

```python
# JALシティ青森のホテル名を取得
jal_city_aomori = soup.find("a", {"title": "ダイワロイネットホテル青森"})
```

```python
print(jal_city_aomori)
```

```python
jal = soup.find_all("p" , attrs={"class":"htlSpecial"})
```

```python
print(jal)
```
